## **7)—Grouped Bar Charts and Box Plots with Plotly**

### **Overview**

After analysing the total number of apps per category and genre, we now investigate the **difference between free and paid apps**, how downloads compare, and whether paid apps can realistically generate meaningful revenue.

This section combines:

* Grouped bar charts
* Box plots
* Log-scaled axes
* Revenue and pricing analysis

---

### **Free vs Paid Apps: Initial Exploration**

```python
df_apps_clean.Type.value_counts()
```

Observation:

* The vast majority of apps on the Play Store are **free**
* Paid apps are relatively rare

To examine this per category:

```python
df_free_vs_paid = df_apps_clean.groupby(
    ['Category', 'Type'],
    as_index=False
).agg({'App': pd.Series.count})

df_free_vs_paid.head()
```

This gives the total number of **Free** and **Paid** apps per category.

---

### **Grouped Bar Chart: Free vs Paid Apps by Category**

```python
g_bar = px.bar(
    df_free_vs_paid,
    x='Category',
    y='App',
    title='Free vs Paid Apps by Category',
    color='Type',
    barmode='group'
)

g_bar.update_layout(
    xaxis_title='Category',
    yaxis_title='Number of Apps',
    xaxis={'categoryorder': 'total descending'},
    yaxis=dict(type='log')
)

g_bar.show()
```

#### Interpretation

* Large categories (Family, Game, Tools) naturally contain more paid apps
* Some smaller categories have a **higher relative share of paid apps**
* Category choice matters if monetisation is the goal

---

### **Question: What Do Paid Apps Give Up?**

Paid apps typically sacrifice downloads. To quantify this, we compare **installs for free vs paid apps**.

---

### **Box Plot: Installs for Free vs Paid Apps**

```python
box = px.box(
    df_apps_clean,
    y='Installs',
    x='Type',
    color='Type',
    notched=True,
    points='all',
    title='How Many Downloads are Paid Apps Giving Up?'
)

box.update_layout(yaxis=dict(type='log'))
box.show()
```

#### Key Insight

* Median installs:

  * **Free apps:** ~500,000
  * **Paid apps:** ~5,000
* Paid apps receive **orders of magnitude fewer downloads**

---

### **Estimating Revenue by Category**

Revenue estimate is based on:

```
Revenue_Estimate = Price × Installs
```

```python
df_paid_apps = df_apps_clean[df_apps_clean['Type'] == 'Paid']

box = px.box(
    df_paid_apps,
    x='Category',
    y='Revenue_Estimate',
    title='How Much Can Paid Apps Earn?'
)

box.update_layout(
    xaxis_title='Category',
    yaxis_title='Paid App Ballpark Revenue',
    xaxis={'categoryorder': 'min ascending'},
    yaxis=dict(type='log')
)

box.show()
```

#### Interpretation

* Most paid apps earn **well below $10,000**
* A small number of **high-earning outliers** exist
* Categories with notable upside:

  * Medical
  * Personalisation
  * Tools
  * Games
  * Family

---

### **Pricing Analysis: How Much Do Paid Apps Cost?**

Median price across all paid apps:

```python
df_paid_apps.Price.median()
```

Result:

* **$2.99**

Now compare pricing by category:

```python
box = px.box(
    df_paid_apps,
    x='Category',
    y='Price',
    title='Price per Category'
)

box.update_layout(
    xaxis_title='Category',
    yaxis_title='Paid App Price',
    xaxis={'categoryorder': 'max descending'},
    yaxis=dict(type='log')
)

box.show()
```

#### Pricing Insights

* Medical apps have the **highest median price** (~$5.49)
* Dating and Business apps are also relatively expensive
* Personalisation apps tend to be cheaper
* Customers in some categories are **less price-sensitive**

---

### **Key Takeaways**

* Free apps dominate downloads; paid apps sacrifice scale
* Most paid apps earn modest revenue
* A small number of outliers drive category profitability
* Pricing tolerance varies significantly by category
* Monetisation strategy must align with category dynamics
