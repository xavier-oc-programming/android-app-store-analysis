## **0)—Day 76 Goals: what you will make by the end of the day**

### **Overview**

In today’s lesson, we analyse **Google Play Store data** and turn raw app metrics into **clear, insightful, and visually appealing charts** using **Plotly**.

The goal is to understand how apps perform across categories, pricing models, and popularity, replicating real-world analytics used by companies such as **App Annie** and **Sensor Tower** to guide product and marketing decisions.

This lesson focuses on **data wrangling, numeric conversion, and high-quality visualisation**, all using Pandas and Plotly.

---

### **Key Questions We Aim to Answer**

* How competitive different app categories are (Games, Lifestyle, Weather, etc.)
* Which app categories offer the strongest opportunities based on popularity
* How many downloads are lost when switching from free to paid apps
* How much you can reasonably charge for a paid app
* Which paid apps generate the highest revenue
* How many paid apps recover their development costs through sales

---

### **Skills You Will Learn Today**

* Removing duplicate rows efficiently with Pandas
* Cleaning data by removing unwanted symbols and converting strings to numeric values
* Handling columns that contain **nested or compound data**
* Creating professional-grade visualisations using the **Plotly** library
* Building:

  * Vertical, horizontal, and grouped bar charts
  * Pie and donut charts for categorical analysis
  * Scatter plots enhanced with colour scales for deeper insight
