## **2)—Preliminary Exploration: The Highest Ratings, Most Reviews, and Largest Size**

### **Identifying the Highest-Rated Apps**

To find the highest-rated apps, we sort the cleaned DataFrame by the `Rating` column in descending order.

```python
df_apps_clean.sort_values('Rating', ascending=False).head()
```

**Observation:**

* Apps with perfect **5.0 ratings** tend to have **very few reviews** and **low install counts**
* This suggests ratings alone are **not a reliable indicator of app quality**
* High ratings with low engagement are often influenced by small sample sizes (e.g. friends and family reviews)

---

### **Finding the Largest Apps by Size**

Next, we inspect which apps are the largest in terms of file size (MB).

```python
df_apps_clean.sort_values('Size_MBs', ascending=False).head()
```

**Observation:**

* The maximum app size observed is **100 MB**
* Multiple apps hit this value exactly, indicating a **hard upper limit**
* This limit is imposed by the **Google Play Store**, not developers
* Apps exceeding this size must rely on additional download mechanisms

---

### **Apps with the Most Reviews**

To identify the most popular apps, we sort by the `Reviews` column and inspect the top 50 entries.

```python
df_apps_clean.sort_values('Reviews', ascending=False).head(50)
```

**Observation:**

* The most-reviewed apps include well-known platforms such as:

  * Facebook
  * WhatsApp
  * Instagram
  * YouTube
* All top 50 most-reviewed apps are **free**
* There are **no paid apps** among the top 50 by number of reviews

---

### **Key Takeaways**

* **Ratings alone can be misleading** without sufficient reviews
* **App size is capped at 100 MB** by Google Play Store policy
* **Popularity (reviews)** is a stronger indicator of success than rating
* Free apps dominate the highest engagement tiers

This preliminary exploration highlights why multiple metrics must be considered together when evaluating app performance.
