## **5)—Plotly Bar Charts & Scatter Plots: The Most Competitive & Popular App Categories**

### **Number of App Categories**

We start by checking how many unique app categories exist.

```python
df_apps_clean.Category.nunique()
```

There are **33 unique categories** in the dataset.

---

### **Most Competitive Categories (Number of Apps)**

To identify which categories are most competitive, we count the number of apps per category.

```python
top10_category = df_apps_clean.Category.value_counts()[:10]
top10_category
```

The most competitive categories (by number of apps) include:

* FAMILY
* GAME
* TOOLS
* PRODUCTIVITY
* PERSONALIZATION

---

### **Bar Chart: Apps per Category**

We visualise the number of apps in each category using a bar chart.

```python
bar = px.bar(
    x=top10_category.index,
    y=top10_category.values
)

bar.show()
```

**Insight:**

* FAMILY and GAME dominate in sheer number of apps
* Entering these categories makes discoverability more difficult due to competition

---

### **Category Popularity (Total Installs)**

Next, we examine popularity by summing installs per category.

```python
category_installs = (
    df_apps_clean
    .groupby('Category')
    .agg({'Installs': pd.Series.sum})
)

category_installs.sort_values(
    'Installs',
    ascending=True,
    inplace=True
)
```

---

### **Horizontal Bar Chart: Category Popularity**

```python
h_bar = px.bar(
    x=category_installs.Installs,
    y=category_installs.index,
    orientation='h',
    title='Category Popularity'
)

h_bar.update_layout(
    xaxis_title='Number of Downloads',
    yaxis_title='Category'
)

h_bar.show()
```

**Insight:**

* GAME and TOOLS categories dominate in total installs
* Popularity does not always correlate with number of apps

---

### **Preparing Data for Scatter Plot**

We create two DataFrames:

* Number of apps per category
* Total installs per category

```python
cat_number = (
    df_apps_clean
    .groupby('Category')
    .agg({'App': pd.Series.count})
)

cat_merged_df = pd.merge(
    cat_number,
    category_installs,
    on='Category',
    how='inner'
)

cat_merged_df.sort_values(
    'Installs',
    ascending=False
)
```

---

### **Scatter Plot: Category Concentration**

We plot category concentration using:

* **x-axis**: Number of apps (competition)
* **y-axis**: Total installs (popularity)
* **Bubble size**: Number of apps
* **Color**: Total installs

```python
scatter = px.scatter(
    cat_merged_df,
    x='App',
    y='Installs',
    title='Category Concentration',
    size='App',
    hover_name=cat_merged_df.index,
    color='Installs'
)

scatter.update_layout(
    xaxis_title='Number of Apps (Lower = More Concentrated)',
    yaxis_title='Installs',
    yaxis=dict(type='log')
)

scatter.show()
```

---

### **Key Insights**

* GAME, FAMILY, and TOOLS are **popular but highly competitive**
* Some categories have **high installs concentrated in few apps**
* Others distribute downloads more evenly across many apps
* Ideal categories balance **popularity and low concentration**
