## **3)—Data Visualisation with Plotly: Create Pie and Donut Charts**

### **Counting Content Ratings**

Each Android app has a content rating (e.g. *Everyone*, *Teen*, *Mature 17+*).
We start by counting how often each rating appears in the dataset.

```python
ratings = df_apps_clean.Content_Rating.value_counts()
ratings
```

This gives us the distribution of content ratings across all apps.

---

### **Importing Plotly Express**

Plotly Express is a high-level interface that allows us to create interactive charts with minimal code.

```python
import pandas as pd
import plotly.express as px
```

---

### **Creating a Basic Pie Chart**

To create a pie chart, we use `px.pie()` and then display the figure with `.show()`.

```python
fig = px.pie(
    labels=ratings.index,
    values=ratings.values
)

fig.show()
```

In Plotly terminology, this creates a **figure (graph object)**.

---

### **Customising the Pie Chart**

We can customise the chart by adding a title and controlling how labels and percentages are displayed.
Additional formatting is done using `.update_traces()`.

```python
fig = px.pie(
    labels=ratings.index,
    values=ratings.values,
    title="Content Rating",
    names=ratings.index
)

fig.update_traces(
    textposition="outside",
    textinfo="percent+label"
)

fig.show()
```

---

### **Creating a Donut Chart**

A donut chart is simply a pie chart with a hole in the centre.
This is controlled using the `hole` parameter.

```python
fig = px.pie(
    labels=ratings.index,
    values=ratings.values,
    title="Content Rating",
    names=ratings.index,
    hole=0.6
)

fig.update_traces(
    textposition="inside",
    textfont_size=15,
    textinfo="percent"
)

fig.show()
```

---

### **Key Takeaways**

* `value_counts()` is a quick way to summarise categorical data
* `plotly.express` allows rapid creation of interactive charts
* Pie charts are useful for **proportion comparisons**
* Donut charts improve readability by focusing attention on percentages
* `.update_traces()` provides fine-grained control over chart appearance
