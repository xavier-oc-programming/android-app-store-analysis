## **4)—Numeric Type Conversions for the Installations & Price Data**

### **Inspecting Data Types**

We begin by checking the data type of the `Installs` column.

```python
df_apps_clean.Installs.describe()
df_apps_clean.info()
```

**Observation:**

* The `Installs` column is stored as an **object**, not a numeric type
* This prevents correct sorting and numerical analysis

---

### **Counting Install Levels (Before Conversion)**

Attempting to group by `Installs` before conversion produces misleading ordering due to commas.

```python
df_apps_clean[['App', 'Installs']].groupby('Installs').count()
```

**Issue:**

* Commas (`,`) cause Python to treat values as strings
* Ordering is lexicographical, not numerical

---

### **Converting `Installs` to Numeric**

This is a two-step process:

1. Remove non-numeric characters
2. Convert the column to a numeric type

```python
df_apps_clean.Installs = (
    df_apps_clean.Installs
    .astype(str)
    .str.replace(',', '')
)

df_apps_clean.Installs = pd.to_numeric(df_apps_clean.Installs)
```

Now grouping by `Installs` produces meaningful numeric ordering.

```python
df_apps_clean[['App', 'Installs']].groupby('Installs').count()
```

**Observations:**

* Only **3 apps** have exactly **1 install**
* **20 apps** have over **1 billion installs**

---

### **Converting the `Price` Column to Numeric**

The `Price` column contains dollar signs and must also be cleaned before analysis.

```python
df_apps_clean.Price = (
    df_apps_clean.Price
    .astype(str)
    .str.replace('$', '')
)

df_apps_clean.Price = pd.to_numeric(df_apps_clean.Price)
```

---

### **Removing Unrealistic Prices**

Apps priced above $250 are removed from the dataset.

```python
df_apps_clean = df_apps_clean[df_apps_clean.Price <= 250]
```

---

### **Estimating App Revenue**

We create a new column estimating revenue as:

[
\text{Revenue_Estimate} = \text{Price} \times \text{Installs}
]

```python
df_apps_clean['Revenue_Estimate'] = (
    df_apps_clean.Price * df_apps_clean.Installs
)
```

---

### **Top-Grossing Paid Apps**

To identify the highest-earning apps:

```python
df_apps_clean.sort_values(
    'Revenue_Estimate',
    ascending=False
).head(10)
```

This ranking shows:

* The **top 10 highest-grossing paid apps**
* A significant portion of these apps belong to the **Games** category

---

### **Key Takeaways**

* Numeric analysis requires explicit type conversion
* Non-numeric characters silently break ordering and grouping
* Install counts vary across many orders of magnitude
* Revenue estimation combines cleaned numeric fields for insight
