## **6)—Extracting Nested Column Data using `.stack()`**

### **Overview**

The `Genres` column is more granular than `Category` and contains **nested values**, where a single app can belong to **multiple genres** separated by semicolons (`;`).
This means we cannot rely on simple `.value_counts()` to get an accurate genre distribution.

---

### **Challenge**

* How many different genres are there really?
* Can an app belong to more than one genre?
* Why does `.value_counts()` give misleading results on nested data?
* How can we fix this using `.split()` and `.stack()`?

---

### **Initial Problem: Nested Values**

```python
len(df_apps_clean.Genres.unique())
```

This returns **114**, but this is misleading.

Using:

```python
df_apps_clean.Genres.value_counts().sort_values(ascending=True).head()
```

Reveals values like:

* `Health & Fitness;Action & Adventure`
* `Arcade;Pretend Play`
* `Card;Brain Games`

This confirms:

* Genres are **semicolon-separated**
* One app can belong to **multiple genres**

---

### **Solution: Split + Stack**

To correctly count genres, we must:

1. Split the strings into separate columns
2. Stack them into a single Series
3. Count the values

```python
stack = df_apps_clean.Genres.str.split(';', expand=True).stack()

print(f"We now have a single column with shape: {stack.shape}")

num_genres = stack.value_counts()
print(f"Number of genres: {len(num_genres)}")
```

Result:

* **53 distinct genres**
* A clean, flat structure suitable for analysis

---

### **Why `.stack()` Matters**

* `.split()` alone creates multiple columns
* `.stack()` collapses them into **one column**
* This allows `.value_counts()` to work correctly

This pattern is essential when dealing with:

* Tags
* Categories
* Multi-label fields
* Any delimited string data

---

### **Key Takeaways**

* Nested string data breaks naive aggregation
* `.split()` + `.stack()` is the canonical Pandas solution
* Always inspect your data before trusting counts
