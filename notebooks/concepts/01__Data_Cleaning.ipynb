## **1)—Data Cleaning: Removing NaN Values and Duplicates**

### **Preliminary Data Exploration**

Before cleaning the data, we first explore what we are working with.

Key inspection questions:

* How many rows and columns does the DataFrame have?
* What are the column names?
* What does a random sample of the data look like?

```python
df_apps.shape
df_apps.columns
df_apps.sample(5)
```

From this inspection, we observe:

* The dataset contains **10,841 rows and 12 columns**
* Some columns contain **NaN values**, especially `Rating` and `Type`
* The `Price` column contains dollar signs (`$`), which will later need cleaning

---

### **Dropping Unused Columns**

Two columns are not needed for the analysis:

* `Last_Updated`
* `Android_Ver`

These are removed using `.drop()` with `axis=1` to specify column removal.

```python
df_apps.drop(['Last_Updated', 'Android_Ver'], axis=1, inplace=True)
```

---

### **Identifying NaN Values**

We focus on rows with missing values in the `Rating` column.

```python
nan_rows = df_apps[df_apps.Rating.isna()]
nan_rows.shape
```

Observations:

* Rows with missing ratings also have **no reviews and no installs**
* This makes these rows unsuitable for meaningful analysis

---

### **Removing NaN Values**

We remove all rows containing NaN values using `.dropna()` and store the cleaned result in a new DataFrame.

```python
df_apps_clean = df_apps.dropna()
df_apps_clean.shape
```

Result:

* **9,367 rows** remain after removing NaN values

---

### **Checking for Duplicates**

Next, we check for duplicate rows using `.duplicated()`.

```python
duplicated_rows = df_apps_clean[df_apps_clean.duplicated()]
duplicated_rows.shape
```

Result:

* **476 duplicate rows** detected

---

### **Investigating Duplicates (Instagram Example)**

To understand duplicates, we inspect entries for a single app.

```python
df_apps_clean[df_apps_clean.App == 'Instagram']
```

Observation:

* Multiple rows exist for the same app
* Differences appear in columns like `Reviews`, making default duplicate removal too strict

---

### **Removing Duplicates Correctly**

To properly identify duplicates, we specify a subset of columns that uniquely define an app.

```python
df_apps_clean = df_apps_clean.drop_duplicates(
    subset=['App', 'Type', 'Price']
)
df_apps_clean.shape
```

Result:

* **8,199 rows** remain after removing duplicates correctly
* Only one unique entry per app (based on name, type, and price) remains

---

### **Important Notes About the Dataset**

* The data is a **sample**, not the full Google Play Store
* It was scraped around **2017–2018**
* Prices are in **USD at the time of scraping**
* App installs are reported in **rounded magnitude ranges** (e.g. `100,000+`)
* The `+` symbol has been removed, and values are treated as exact for simplicity
* App size is represented in **MB**, with missing values replaced by category averages

---

This cleaned dataset is now suitable for further numerical conversion and visualisation.
