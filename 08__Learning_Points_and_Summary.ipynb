## **8)—Learning Points & Summary**

### **What We Covered in This Lesson**

* Pulling a **random sample** from a DataFrame using `.sample()`
* Identifying duplicate entries with `.duplicated()`
* Removing duplicates using `.drop_duplicates()`
* Converting string / object columns to numeric data with `.to_numeric()`
* Using **Plotly** to create:

  * Pie charts
  * Donut charts
  * Bar charts
  * Box plots
  * Scatter plots

---

### **Core Data Skills Reinforced**

* Inspecting and validating data types before analysis
* Cleaning real-world datasets with inconsistent formatting
* Handling nested and non-numeric values safely
* Choosing appropriate visualisations for comparison and distribution analysis
* Using log scales to reveal patterns hidden by skewed data

---

### **Key Takeaway**

By combining **data cleaning**, **type conversion**, and **interactive visualisation**, we can extract meaningful insights from scraped Google Play Store data and build a much clearer picture of how the app ecosystem behaves across categories, pricing models, and monetisation strategies.
